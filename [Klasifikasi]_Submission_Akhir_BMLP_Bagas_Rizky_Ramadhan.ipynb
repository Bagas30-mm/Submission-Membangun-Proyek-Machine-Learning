**Bagaimana Algoritma Ini Bekerja:**

**langkah pertama: Pendekatan Kolektif (*Ensemble Learning*):**  
Random Forest merupakan bagian dari metode ensemble, yaitu teknik yang menggabungkan hasil dari banyak model sederhana (dalam hal ini pohon keputusan) untuk mendapatkan prediksi yang lebih kuat dan stabil. Ibaratnya seperti meminta pendapat dari banyak orang untuk mengambil keputusan terbaik, bukan hanya dari satu orang saja.

**Langkah kedua: Pembuatan Pohon Keputusan yang Beragam:**

- Algoritma ini membuat banyak *decision tree* (pohon keputusan).
- Setiap pohon dilatih menggunakan data yang diambil secara acak dari dataset asli melalui teknik *bootstrap sampling* — yaitu pengambilan sampel dengan pengembalian. Sehingga, setiap pohon belajar dari subset data yang berbeda-beda.
- Saat membentuk struktur pohon, hanya sebagian fitur yang dipilih secara acak di tiap titik percabangan. Teknik ini disebut *feature randomness*, dan bertujuan agar tiap pohon memiliki pola yang unik. Dengan begitu, model menjadi lebih tangguh dan tidak mudah overfitting terhadap data pelatihan.

### **Langkah Ketiga: Proses Prediksi**

Masing-masing pohon akan memberikan hasil prediksinya secara independen.  
Untuk tugas klasifikasi seperti pada kasus ini, keputusan akhir diambil berdasarkan hasil *voting* mayoritas—yaitu kelas yang paling banyak dipilih oleh pohon-pohon tersebut akan dijadikan sebagai hasil prediksi model.

### **Istilah dalam Classification Report**

- **Precision**: Mengukur seberapa tepat prediksi model terhadap kelas tertentu. Artinya, dari semua prediksi yang menyatakan "positif", berapa banyak yang benar-benar sesuai.
  - **Rumus**: `Precision = True Positive / (True Positive + False Positive)`

- **Recall**: Menilai kemampuan model dalam menemukan semua data yang benar-benar positif. Semakin tinggi nilai ini, semakin baik model dalam mendeteksi data yang seharusnya terklasifikasi.
  - **Rumus**: `Recall = True Positive / (True Positive + False Negative)`

- **F1-Score**: Merupakan rata-rata harmonis antara precision dan recall, berguna untuk menyeimbangkan keduanya terutama saat data tidak seimbang.
  - **Rumus**: `F1-Score = 2 * (Precision * Recall) / (Precision + Recall)`

- **Support**: Jumlah total data aktual untuk masing-masing kelas dalam dataset uji.

#### **macro avg**:
